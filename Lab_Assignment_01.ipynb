```python
# Load ratings and movies datasets
ratings = pd.read_csv("/content/drive/MyDrive/DM and W/ratings.csv")
movies = pd.read_csv("/content/drive/MyDrive/DM and W/movies.csv")

# Create User-Movie Rating Matrix
user_movie_matrix = ratings.pivot(index="userId", columns="movieId", values="rating")

# Fill missing values with 0 (or use mean imputation)
user_movie_matrix = user_movie_matrix.fillna(0)

# Compute Movie Similarity Matrix using Pearson Correlation
movie_similarity = user_movie_matrix.corr(method="pearson")
```

```python
# Function to recommend similar movies
def recommend_movies(movie_id, top_n=5):
    if movie_id not in movie_similarity.index:
        return f"Movie ID {movie_id} not found in similarity matrix."

    # Get top N similar movies
    similar_movies = movie_similarity[movie_id].dropna().sort_values(ascending=False)[1:top_n+1]

    # Merge with movie titles
    recommended = movies[movies["movieId"].isin(similar_movies.index)][["movieId", "title"]].copy()

    # Add similarity scores
    recommended["similarity"] = recommended["movieId"].map(similar_movies)

    # Sort by similarity score
    recommended = recommended.sort_values(by="similarity", ascending=False).reset_index(drop=True)

    return recommended

    movie_id = 3
recommendations = recommend_movies(3, top_n=5)
print(recommendations)
```

```python
#Q 01: Select a specific user from the dataset
user_ratings = ratings[ratings["userId"] == user_id]
print("User Ratings Data:\n", user_ratings)
```

```python
#Q 02: Identify the movie the user has rated the highest
if not user_ratings.empty:
    highest_rated_item = user_ratings.loc[user_ratings["rating"].idxmax()]
    print("\nHighest Rated Movie by User:\n", highest_rated_item)
else:
    print("\nUser has not rated any movies.")
```

```python
#Q 03: Find movies that the user has NOT rated yet
all_movies = set(movies["movieId"])  # All available movies
rated_movies = set(user_ratings["movieId"])  # Movies rated by the user
unrated_movies = all_movies - rated_movies  # Movies the user has NOT rated

# Get details of unrated movies
unrated_movies_details = movies[movies["movieId"].isin(unrated_movies)]
print("\nUnrated Movies:\n", unrated_movies_details.head())
```

```python
#Q 04: Recommend movies with the highest average rating
def recommend_movies_with_highest_avg_rating(user_id, top_n=5):

    # Get movies rated by the user
    user_ratings = ratings[ratings["userId"] == user_id]
    rated_movies = set(user_ratings["movieId"])  # Movies the user has rated
    all_movies = set(movies["movieId"])  # All available movies

    # Find movies the user has NOT rated
    unrated_movies = all_movies - rated_movies
    unrated_movies_details = movies[movies["movieId"].isin(unrated_movies)]

    # Compute average rating for each movie
    avg_ratings = ratings.groupby("movieId")["rating"].mean()

    # Merge unrated movies with their average ratings
    recommended_movies = unrated_movies_details.merge(
        avg_ratings, on="movieId", how="left"
    ).sort_values(by="rating", ascending=False).head(top_n)

    return recommended_movies[["movieId", "title", "rating"]]

# Get top 5 recommended movies
recommended_movies = recommend_movies_with_highest_avg_rating(user_id=47, top_n=5)
print("\nTop Recommended Movies:\n", recommended_movies)
```
